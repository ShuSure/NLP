## NLP

```python
punc = ['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}',"%"]
stop_words = text.ENGLISH_STOP_WORDS.union(punc)
desc = variety_df['description'].values
vectorizer = TfidfVectorizer(stop_words = stop_words)
X = vectorizer.fit_transform(desc)
```

```python
word_features = vectorizer.get_feature_names()
word_features[550:575]
```

```python
stemmer = SnowballStemmer('english')
tokenizer = RegexpTokenizer(r'[a-zA-Z\']+')

def tokenize(text):
    return [stemmer.stem(word) for word in tokenizer.tokenize(text.lower())]
```

```python
vectorizer2 = TfidfVectorizer(stop_words = stop_words, tokenizer = tokenize)
X2 = vectorizer2.fit_transform(desc)
word_features2 = vectorizer2.get_feature_names()
word_features2[:50]
```

```python
vectorizer3 = TfidfVectorizer(stop_words = stop_words, tokenizer = tokenize, max_features = 1000)
X3 = vectorizer3.fit_transform(desc)
words = vectorizer3.get_feature_names()
```

```python
kmeans = KMeans(n_clusters = 10, n_init = 5, n_jobs = -1)
kmeans.fit(X3)
```

```python
common_words = kmeans.cluster_centers_.argsort()[:,-1:-11:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' : ' + ', '.join(words[word] for word in centroid))
```

```python
df.dropna()
df.drop(columns = ['Unnamed: 0','country',
                   'points','price',
                   'region_2','taster_name','taster_twitter_handle','title'],axis=1,inplace = True) #remove useless cols
```

```python
df.replace("nan",np.nan,inplace=True)
df=df.dropna()
df['cluster']=df['cluster'].apply(str)
```

```python
df.loc[(df['winery']=='Three Sticks')]
```

```python
def content_based_recommender(winery):
    
    winery = str(winery)
    if winery in df['winery'].values:
        common_wineries = df
        common_wineries = common_wineries.drop_duplicates(subset=['winery'])
        common_wineries.reset_index(inplace= True)
        common_wineries['index'] = [i for i in range(common_wineries.shape[0])]
        target_cols = ['region_1','province','variety','cluster','description']
        common_wineries['combined_features'] = [' '.join(common_wineries[target_cols].iloc[i,].values) for i in range(common_wineries[target_cols].shape[0])]
        cv = CountVectorizer()
        count_matrix = cv.fit_transform(common_wineries['combined_features'])
        cosine_sim = cosine_similarity(count_matrix)
        index = common_wineries[common_wineries['winery'] == winery]['index'].values[0]
        sim_wineries = list(enumerate(cosine_sim[index]))
        sorted_sim_wineries = sorted(sim_wineries,key=lambda x:x[1],
                                      reverse=True)[1:6]
            
        wineries = []
        for i in range(len(sorted_sim_wineries)):
            wineries.append(common_wineries[common_wineries['index'] == sorted_sim_wineries[i][0]]['winery'].item())
        return(wineries)
                     
    else:
        
        return('Cant find winery in dataset, please check spelling')
```

```python
content_based_recommender('Ponzi')
```

```python
app = Flask(__name__)

@app.route("/")
def home():
    return render_template('home.html')

@app.route("/recommend")
def recommend():
    winery = request.args.get('winery')
    rec = content_based_recommender(winery)
    wnery = winery.upper()
    if type(rec)==type('string'):
        return render_template('recommend.html',winery=winery,rec=rec,t='s')
    else:
        return render_template('recommend.html',winery=winery,rec=rec,t='wineries')



if __name__ == '__main__':
    app.run()
```
